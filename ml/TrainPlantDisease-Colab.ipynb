# Plant Disease Training (Colab)

This notebook trains your project model with a stronger backbone and exports TensorFlow.js files.

Run all cells in order.


```python
import os
import subprocess

# Set this if you want to clone directly from GitHub.
# Leave empty if you already uploaded the project at PROJECT_PATH.
REPO_URL = ""
PROJECT_PATH = "/content/Farm-Connnect"

if not os.path.exists(PROJECT_PATH):
    if REPO_URL:
        subprocess.run(["git", "clone", REPO_URL, PROJECT_PATH], check=True)
    else:
        raise RuntimeError(
            "Project folder not found. Set REPO_URL or upload your project to /content/Farm-Connnect"
        )

print("Project path:", PROJECT_PATH)
```

```python
os.chdir(PROJECT_PATH)
subprocess.run(["python", "-m", "pip", "install", "--upgrade", "pip"], check=True)
subprocess.run(["python", "-m", "pip", "install", "-r", "ml/requirements.txt"], check=True)
print("Dependencies installed")
```

```python
import tensorflow as tf
print("TensorFlow:", tf.__version__)
print("GPU devices:", tf.config.list_physical_devices("GPU"))
```

```python
# Training config
BACKBONE = "efficientnetb0"  # or "mobilenetv2"
EPOCHS = 20
FINE_TUNE_EPOCHS = 12
BATCH_SIZE = 16
UNFREEZE_LAYERS = 80
OUTPUT_DIR = "ml/saved_model_colab"
DENSE_UNITS = "384"
DROPOUT = "0.35"
```

```python
train_cmd = [
    "python", "ml/train.py",
    "--backbone", BACKBONE,
    "--epochs", str(EPOCHS),
    "--fine-tune-epochs", str(FINE_TUNE_EPOCHS),
    "--batch-size", str(BATCH_SIZE),
    "--unfreeze-layers", str(UNFREEZE_LAYERS),
    "--output-dir", OUTPUT_DIR,
    "--dense-units", DENSE_UNITS,
    "--dropout-rate", DROPOUT,
]

print("Running:", " ".join(train_cmd))
subprocess.run(train_cmd, check=True)
```

```python
EXPORT_DIR = "ml/tfjs_model_colab"
export_cmd = [
    "python", "ml/export.py",
    "--model", f"{OUTPUT_DIR}/plant_disease_model.h5",
    "--output", EXPORT_DIR,
]

print("Running:", " ".join(export_cmd))
subprocess.run(export_cmd, check=True)

import shutil
shutil.make_archive("ml/tfjs_model_colab", "zip", root_dir=".", base_dir=EXPORT_DIR)
print("Created: ml/tfjs_model_colab.zip")
```
